## Univariate Analysis

```python
categorical_column=data.select_dtypes(include=['object', 'category']).columns
data[categorical_column].apply(lambda x : len(x.unique()))
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(data['country_txt'].value_counts()[:20].index,data['country_txt'].value_counts()[:20].values)
plt.title('Top 20 countries of terrorist attack')
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(data['city'].value_counts()[1:21].index,data['city'].value_counts()[1:21].values)
plt.title('Top 20 cities of terrorist attack')
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(data['region_txt'].value_counts().index,data['region_txt'].value_counts().values)
plt.title('Region wise Terrorist Attack')
plt.show()
```

```python
                
plt.subplots(figsize=(25,7))
sns.barplot(data['target1'].value_counts()[0:20].index,data['target1'].value_counts()[0:20].values)
plt.title('Most Targeted Places')
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(data['attacktype1_txt'].value_counts().index,data['attacktype1_txt'].value_counts().values)
plt.show()
```

## Analysis Categorical vs Categorical

```python
plt.figure(figsize=(20,20))
cross_tab = pd.crosstab(data['iyear'],data['attacktype1_txt'],margins=True)
cross_tab.iloc[:-1,:-1].plot(kind='area',stacked=False, color=['red','blue','green','purple','orange','seagreen','pink','yellow','brown'], grid=False,figsize=(15,6))
plt.title('Year wise attack type')
plt.show()
```

```python
fig = plt.subplots(figsize=(15,6))

sns.lineplot(data[data['country_txt']=='Pakistan']['iyear'].value_counts().index,
             data[data['country_txt']=='Pakistan']['iyear'].value_counts().values)
sns.lineplot(data[data['country_txt']=='India']['iyear'].value_counts().index,
             data[data['country_txt']=='India']['iyear'].value_counts().values)
sns.lineplot(data[data['country_txt']=='Afghanistan']['iyear'].value_counts().index,
             data[data['country_txt']=='Afghanistan']['iyear'].value_counts().values)
sns.lineplot(data[data['country_txt']=='Iraq']['iyear'].value_counts().index,
             data[data['country_txt']=='Iraq']['iyear'].value_counts().values)
plt.legend(labels=['Pakistan','India','Afghanistan','Iran'])
plt.show()
```
