## Univariate Analysis

```python
categorical_column=deliveries.select_dtypes(include=['object', 'category']).columns
deliveries[categorical_column].apply(lambda x : len(x.unique()))
```

```python
categorical_column=matches.select_dtypes(include=['object', 'category']).columns
matches[categorical_column].apply(lambda x : len(x.unique()))
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(matches['winner'].value_counts().index,matches['winner'].value_counts().values)
plt.title('Top winner teams')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(matches['player_of_match'].value_counts()[0:10].index,matches['player_of_match'].value_counts()[0:10].values)
plt.title('Top players in IPL')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(25,7))
deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar',color='black') 
plt.title('Most run  by the batsman')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(matches['venue'].value_counts()[0:10].index,matches['venue'].value_counts()[0:10].values)
plt.title('Most venues consider in IPL')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(25,7))
sns.barplot(matches['city'].value_counts().index,matches['city'].value_counts().values)
```

```python
plt.subplots(figsize=(25,7))
group=matches.groupby('venue')['city'].count().sort_values(ascending=False)
sns.barplot(group.index, group.values) #(month, day)
plt.title('Most matches in the city')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(25,7))
plt.pie(matches['toss_decision'].value_counts(), labels=matches['toss_decision'].value_counts().index, autopct='%1.1f%%', startangle=180)
plt.show()
```

```python
plt.subplots(figsize=(35,7))
sns.barplot(matches[matches['toss_decision']=='field']['winner'].value_counts().index,
             matches[matches['toss_decision']=='field']['winner'].value_counts().values)
plt.show()
```
